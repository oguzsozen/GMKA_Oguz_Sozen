1.  3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir pivot tablo üretiniz.

```python
import yfinance as yf
import plotly.express as px
import pandas as pd
```

```python
B030 = yf.download("XU030.IS", start="2021-01-01", end="2022-12-31")
Kur = yf.download("TRY=X", start="2021-01-01", end="2022-12-31")
```

```python
B030 = B030[["Adj Close"]]
B030.columns = ["Kapanış (TRY)"]
B030["Normalize"] = B030["Kapanış (TRY)"]/B030.iloc[0,0]

Kur = Kur[["Adj Close"]]
Kur.columns = ["USD/TRY"]
Kur["Normalize USD/TRY"] = Kur["USD/TRY"]/Kur.iloc[0,0]
```

```python
B030K = B030.merge(Kur, left_index=True, right_index=True)
B030K
```

```python
B030K.describe()
```

```python
B030K.info()
```

```python
B030K["Ay_Yıl"] = B030K.index.strftime("%Y-%m")
```

```python
B030K["Ay_Yıl"] = B030K.index.strftime("%Y-%m")
B030K_as = B030K.pivot_table(values=["Kapanış (TRY)","USD/TRY"], index="Ay_Yıl", aggfunc="last")
B030K_as
```

```python
B030K_as["Değişim"] = B030K_as["Kapanış (TRY)"].diff(1)
B030K_as["D_Yön"] = B030K_as["Değişim"].map(lambda x: "Düşen Gün" if x < 0 else "Yükselen Gün")
B030K_as["Değişim %"] = (B030K_as["Değişim"]/B030K_as["Kapanış (TRY)"].shift(1))
B030K_as
```

2. Ay kırılımlı olarak ürettiğiniz tabloyu grafik olarak inceleyiniz.


```python
px.bar(B030K_as, y="Değişim %",
       color= "D_Yön",
       color_discrete_map={"Düşen Gün": "red", "Yükselen Gün":"green"},
       text_auto=".0%")
```

3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir tabloyu bir de resample yöntemi ile üretiniz.

```python
B030K_as2 = B030K.resample("BM", convention="end").asfreq()[["Kapanış (TRY)","USD/TRY"]]
B030K_as2
```

# Null olan aylara bir önceki ve sonraki ayların aritmetik ortalaması olacak şekilde yerleştirildi.

```python
B030K_as2.iloc[4,0] = (B030K_as2.iloc[3,0]+B030K_as2.iloc[5,0])/2
B030K_as2.iloc[4,1] = (B030K_as2.iloc[3,1]+B030K_as2.iloc[5,1])/2
B030K_as2.iloc[9,0] = (B030K_as2.iloc[8,0]+B030K_as2.iloc[10,0])/2
B030K_as2.iloc[9,1] = (B030K_as2.iloc[8,1]+B030K_as2.iloc[10,1])/2
B030K_as2
```

```python
B030K_as2["Değişim"] = B030K_as2["Kapanış (TRY)"].diff(1)
B030K_as2["D_Yön"] = B030K_as2["Değişim"].map(lambda x: "Düşen Gün" if x < 0 else "Yükselen Gün")
B030K_as2["Değişim %"] = (B030K_as2["Değişim"]/B030K_as2["Kapanış (TRY)"].shift(1))
B030K_as2
```

4. Resample yöntemi ile ürettiğiniz tabloyu grafik olarak inceleyiniz. 2. adımda elde ettiğiniz grafik ile karşılaştırarak yorumlayınız.

```python
px.bar(B030K_as2, y="Değişim %",
       color= "D_Yön",
       color_discrete_map={"Düşen Gün": "red", "Yükselen Gün":"green"},
       text_auto=".0%")
```
