### BIST30 Endeksinin incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait BIST30 endeks değerlerini indiriniz

```python
import yfinance as yf
import plotly.express as px
```

```python
B030 = yf.download("XU030.IS", start="2021-01-01", end="2021-12-31")
B030
```

2. Veriyi özet metotları ile inceleyiniz (head, info, describe)

```python
B030.head(10)
```

```python
B030.info()
```

```python
B030.describe()
```

3. Veri temizleme:<br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
B030 = B030[["Adj Close"]]
B030
```

```python
B030.columns = ["Kapanış (TRY)"]
B030
```

4. Tabloya yeni sütun ekleme:<br>
*   Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz (yani tüm değerleri 1 Ocak 2021 değerine bölünüz). Bu alana "Normalize" alanı diyebilirsiniz.<br>



```python
norm = B030.iloc[0,0]
B030["Normalize"] = B030["Kapanış (TRY)"]/norm
B030
```

5. Grafik çizdirme ve yorumlama:
*   Tabloyu px.line grafiği olarak görselleştiriniz.<br>
*   BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

# *BIST 030 endeksi 2021 yılında Mart ayının sonlarına kadar genellikle 1'in hemen altında seyrederken Mart sonlarından Kasım ayının ortalarına kadar ise 0.86 ile 0.95 değerleri aralığında dalgalanmalar yaşamaktadır. Buradan yükselişe geçen BIST 030 Aralık ayının ortalarına kadar çok hızlı ve istikrarlı bir yükseliş gösterip 1.53 seviyelerini yakalamayı başarmış olmasına rağmen ulaştığı en yüksek noktadan 1 hafta içerisinde çok sert bir düşüş yaparak 1.25 civarlarında dengelenmiştir.*

```python
px.line(B030, y="Normalize",
        title="BIST03O Endeksi 2021 - Normalize (1 Ocak 2021)",
        hover_data={"Normalize" : ":.2f", "Kapanış (TRY)":":.2f"},
        labels={"Date":"Tarih","Normalize":"BIST 030 Endeksi (TRY)"})
```

6. Değişim alanlarını ekleme ve grafikte koşullu renklendirme:<br>
	• Normalize etiğiniz alan için her satırda bir önceki satırdaki Normalize değeri ile olan farkını hesaplattırınız (bu sütunu "degisim" olarak isimlendirebilirsiniz).
	• Tabloyu px.bar grafiği olarak görselleştiriniz.<br>
	• px.bar grafiğinde endeksin yükseldiği günleri "yeşil", düştüğü günleri "kırmızı" olarak görselleştiriniz.

```python
B030["degisim"] = B030["Normalize"].diff(1)
B030
```

```python
B030["D_Yön"] = B030["degisim"].map(lambda x: "Düşen Gün" if x < 0 else "Yükselen Gün")
B030
```

```python
B030["degisim %"] = (B030["degisim"]/B030["Normalize"].shift(1))*100
B030
```

```python
px.bar(B030, y="degisim %",
       color= "D_Yön",
       color_discrete_map={"Düşen Gün": "red", "Yükselen Gün":"green"})
```

7. Histogram çizmek:

*   Bölüm sayısını (nbins) nasıl belirlediniz,açıklayınız.
*   Elde ettiğiniz grafiği yorumlayınız.



# *Senin çoğunda endeks +-%2 kadar günlük değişim göstermiştir. Sadece 3 gün için endekste %8'den daha fazla bir düşüş görülmüştür.*

```python
px.histogram(B030, "degisim %", nbins=8) 

# nbins için grafiğin en genel dağılımını kaybetmeyen ve sonuçlardan anlamlı
# gruplamalar yapılabilen değer 8 olarak belirlenmiştir.
```
